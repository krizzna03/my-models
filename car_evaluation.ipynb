```python
from sklearn.preprocessing import OrdinalEncoder
```

```python
oe = OrdinalEncoder()

x['buying'] = oe.fit_transform(x[['buying']])
x['maint'] = oe.fit_transform(x[['maint']])
x['doors'] = oe.fit_transform(x[['doors']])
x['persons'] = oe.fit_transform(x[['persons']])
x['lug_boot'] = oe.fit_transform(x[['lug_boot']])
x['safety'] = oe.fit_transform(x[['safety']])
```

```python
x = x.iloc[:, 0:-1]
y = x.iloc[:, -1]
```

```python
from sklearn.model_selection import train_test_split
```

```python
x_train, x_test, y_train, y_test = train_test_split(x, y, test_size = 0.25, random_state = 0)
```

```python
from sklearn.ensemble import RandomForestClassifier

rfc = RandomForestClassifier()

rfc.fit(x_train, y_train)
```

```python
pred1 = rfc.predict(x_test)
pred1
```

```python
from sklearn.metrics import accuracy_score

accuracy_score(y_test, pred1)
```
