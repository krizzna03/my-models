```python
from sklearn.linear_model import LogisticRegression
```

```python
labels = ['age_first_funding_year', 'age_last_funding_year' , 'funding_rounds', 'funding_total_usd', 'age_first_milestone_year',
          'age_last_milestone_year', 'is_top500', 'has_VC']
x = df[labels]
y = df['status']
```

```python
from sklearn.model_selection import train_test_split
x_train, x_test, y_train, y_test = train_test_split(x, y, test_size = 0.3, random_state = 42)

model = LogisticRegression()
model.fit(x_train, y_train)
```

```python
y_pred = model.predict(x_test)
from sklearn.metrics import classification_report as cr, accuracy_score as ac
```

```python
print(cr(y_pred, y_test))
print(ac(y_pred, y_test))
```

```python
num_cols = x.select_dtypes(include=['float64', 'int64']).columns
```

```python
imputer = SimpleImputer(strategy='mean')
x_train[num_cols] = imputer.fit_transform(x_train[num_cols])
x_test[num_cols] = imputer.transform(x_test[num_cols])
```

```python
from sklearn.preprocessing import StandardScaler
sc = StandardScaler()
x_train[num_cols] = sc.fit_transform(x_train[num_cols])
x_test[num_cols] = sc.transform(x_test[num_cols])
```

```python
model2 = LogisticRegression()
model2.fit(x_train, y_train)
y_pred2 = model2.predict(x_test)
print(cr(y_pred2, y_test))
print(ac(y_pred2, y_test))
```

```python
from sklearn.ensemble import RandomForestClassifier
rfc = RandomForestClassifier(class_weight = 'balanced', random_state = 0)
rfc.fit(x_train, y_train)
y_rfc = rfc.predict(x_test)
print(cr(y_rfc, y_test))
print(ac(y_rfc, y_test))
```
