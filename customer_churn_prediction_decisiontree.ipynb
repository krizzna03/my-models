```python
import pandas as pd 
import matplotlib.pyplot as plt
from sklearn.tree import DecisionTreeClassifier, export_text, plot_tree
from sklearn.metrics import accuracy_score, classification_report
from sklearn.model_selection import train_test_split
```

```python
data = pd.read_csv("churn.csv")
data.head(5)
```

```python
x = data.drop(columns=['Churn','customerID','gender'])
x_enc = pd.get_dummies(x, drop_first=True)
y = data["Churn"].map({"No": 0, "Yes": 1}) 
```

```python
x_train, x_test, y_train, y_test = train_test_split(x_enc, y, test_size = 0.30, random_state =42, stratify=y)
```

```python
clf = DecisionTreeClassifier(criterion = 'gini', max_depth = 4, random_state = 42)
clf.fit(x_train, y_train)
```

```python
y_pred = clf.predict(x_test)
print("Accuracy Score:", accuracy_score(y_test, y_pred))
print("Classification report:", classification_report(y_test, y_pred))
```

```python
print("Gini Decision tree:")
print(export_text(clf, feature_names = list(x_enc.columns)))
```

```python
plt.figure(figsize=(80,40))
plot_tree(clf, feature_names = x_enc.columns, class_names = ['No', 'Yes'], filled=True)
plt.show()
```
